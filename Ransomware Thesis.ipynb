```python
import pandas as pd
df = pd.read_csv('Ransomware.csv', sep='|')
df
```

```python
count_0 = df[df['legitimate'] == 0].shape[0]
count_1 = df[df['legitimate'] == 1].shape[0]

print(f"Count of 0 in 'legitimate': {count_0}")
print(f"Count of 1 in 'legitimate': {count_1}")
```

```python
feats = df[['SizeOfOptionalHeader', 'MajorLinkerVersion', 'AddressOfEntryPoint', 'SectionAlignment', 'MinorOperatingSystemVersion', 'SizeOfHeaders', 'SizeOfStackReserve', 'LoaderFlags', 'SectionsMinEntropy', 'SectionsMaxEntropy', 'SectionMaxRawsize', 'SectionsMinVirtualsize', 'ResourcesMinEntropy', 'legitimate']]
```

```python
feats
```

```python
X = feats.drop(columns=['legitimate'])
y = feats['legitimate']
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
print(type(X_test))
count_0 = y_test[df['legitimate'] == 0].shape[0]
count_1 = y_test[df['legitimate'] == 1].shape[0]

print(f"Count of 0 in 'legitimate': {count_0}")
print(f"Count of 1 in 'legitimate': {count_1}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
#H-ML CELL : combining a heuristic with machine learning 
import time
startTime = time.time()
heuristicCaught = 0
heuristicCorrect = 0

hybridPred = np.zeros(len(X_test))

for i in range(0, len(X_test.index)):
    currRow = X_test.iloc[[i]]
    if ((X_test.iloc[i]['SectionsMaxEntropy'] + X_test.iloc[i]['SectionsMinEntropy'])/2 > 7): #HEURISTIC HERE
        heuristicCaught += 1
        hybridPred[i] = 0
        if y_test.iloc[i] == 0:
            heuristicCorrect += 1
    else:
        hybridPred[i] = dt.predict(currRow) #ML MODEL HERE
endTime = time.time()
runTime = endTime - startTime
```

```python
#going to need to find a balance...higher average threshold i set, less caught by heuristic but heuristic more correct
#need to find an average entropy from cybersecurity world to use
print("HEURISTIC-ML HYBRID DATA")
print(classification_report(y_test, hybridPred))
print(confusion_matrix(y_test, hybridPred))
print("Flagged by heuristic: " + str(heuristicCaught))
print("Correct flaggings: " + str(heuristicCorrect))
print("Heuristic False Positives: " + str(heuristicCaught - heuristicCorrect))
print("Percent of correct flags made by heuristic out of total ransomware: " + str((heuristicCorrect / 28942) * 100))
print("Runtime: " + str(runTime))
```

```python
#PURE ML CELL : just machine learning 
import time
startTime = time.time()

purePred = np.zeros(len(X_test))

for i in range(0, len(X_test.index)):
    currRow = X_test.iloc[[i]]
    purePred[i] = dt.predict(currRow) #ML MODEL HERE
    
endTime = time.time()
runTime = endTime - startTime
```

```python
print("PURE ML DATA")
print(classification_report(y_test, purePred))
print(confusion_matrix(y_test, purePred))
print("Runtime: " + str(runTime))
```
